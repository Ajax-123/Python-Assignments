# Problem 1:

Scipy:
We have the min and max temperatures in a city In India for each months of the year. 
We would like to find a function to describe this and show it graphically, the dataset 
given below.
Task:
1.fitting it to the periodic function
2.plot the fit
Data
Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25
Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize

temp_min= np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])

temp_max= np.array([ 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])

mon = np.arange(12)
```

```python
def yearly_temp(times,avg,amp1,time_offset):
    return(avg+amp1*np.cos((times+time_offset)*2*np.pi/times.max()))

res_max,cov_max=optimize.curve_fit(yearly_temp,mon,temp_max,[40,20,0])

res_min,cov_min=optimize.curve_fit(yearly_temp,mon,temp_min,[-40,20,0])
```

```python
days= np.linspace(0,12,num=365)
```

```python
plt.figure()
plt.plot(mon,temp_max,'ro')
plt.plot(days,yearly_temp(days,*res_max),'r-')
plt.plot(mon,temp_min,'bo')
plt.plot(days,yearly_temp(days,*res_min),'b-')
plt.xlabel('Month')
plt.ylabel('Temperature')
plt.show()
```

```python
import pandas as pd
import numpy as no
import matplotlib.pyplot as plt
```

```python
url='https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
```

```python
Titanic = pd.read_csv(url)
Titanic
```

#1. Create a pie chart presenting the male/female proportion

```python
counts=Titanic['sex'].value_counts()
fig, axis1 = plt.subplots(ncols=1, figsize=(5, 5))
axis1.pie(counts.values.tolist(),labels=counts.index.values.tolist(),colors=['blue','red'],startangle=90, autopct='%.1f%%')
axis1.set_title('Male-female proportion')
```

#2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
plt.figure() 
s = Titanic[Titanic.sex=='male'].plot.scatter('age', 'fare', color='blue',label='Male') 
Titanic[Titanic.sex=='female'].plot.scatter('age', 'fare',color='red',label='Female',ax=s)
```
