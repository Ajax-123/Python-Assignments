In this assignment students have to transform iris data into 3 dimensions 
and plot a 3d chart with transformed dimensions and colour each data 
point with specific class.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
from sklearn.decomposition import PCA
%matplotlib inline
```

```python
iris = datasets.load_iris()
```

```python
classes = PCA(n_components=3).fit_transform(iris.data)
#classes=iris.data[:,:-1]
target=iris.target
```

```python
figure=plt.figure(1,figsize=(8,6))
axis=Axes3D(figure,elev=-150,azim=110)
axis.scatter(classes[:,0],classes[:,1],classes[:,2],c=target,cmap=plt.cm.rainbow_r,alpha=0.9,s=50)
axis.set_title("Iris 3D Chart")
axis.set_xlabel("X-axis")
axis.set_ylabel("Y-axis")
axis.set_zlabel("Z-axis")

plt.show()
```
